```python
from supabase import create_client, Client
import pandas as pd
from dotenv import load_dotenv
import os
```

```python
# Load the CSV data
file_path = 'BSE Equity.csv'
data = pd.read_csv(file_path)
```

```python
# Select the relevant columns
filtered_data = data[['Security Code', 'Security Id', 'Security Name', 'ISIN No', 'Industry', 
                      'Sector Name', 'Industry New Name', 'Igroup Name', 'ISubgroup Name']]
```

```python
# Replace null values with empty strings
filtered_data = filtered_data.fillna('Others')
```

```python
load_dotenv()
# Supabase credentials
url = os.getenv('SUPABASE_URL')
key = os.getenv('SUPABASE_KEY')
```

```python
# Initialize Supabase client
supabase: Client = create_client(url, key)
```

```python
# Insert data into Supabase
for index, row in filtered_data.iterrows():
    response = supabase.table('securities').insert({
        'security_id': row['Security Id'],
        'security_code': row['Security Code'],
        'security_name': row['Security Name'],
        'isin_no': row['ISIN No'],
        'industry': row['Industry'],
        'sector_name': row['Sector Name'],
        'industry_new_name': row['Industry New Name'],
        'igroup_name': row['Igroup Name'],
        'isubgroup_name': row['ISubgroup Name']
    }).execute()

    # Print response for each row to see success/failure
    print(response)
```

```python
# get the item with specific security_id form the table
response = supabase.table('securities').select().eq('security_name', 'Borosil Renewables Ltd').execute()

```

```python
res = requests.get("https://analyst.indianapi.in/static/all_stocks.json")
stocks = res.json()
```

```python
for stock in stocks:
    if stock["nse-code"] != "" and stock["bse-code"] == "":
         # Check if stock already exists by security_id (NSE code)
        existing = supabase.table('securities').select('security_id').eq('security_id', stock['nse-code']).execute()
        
        if not existing.data:
            # Insert new NSE stock
            response = supabase.table('securities').insert({
                'security_id': stock['nse-code'],
                'security_code': stock['nse-code'],
                'security_name': stock['name'],
                'exchange': 'NSE'
            }).execute()

            print(response)
```
